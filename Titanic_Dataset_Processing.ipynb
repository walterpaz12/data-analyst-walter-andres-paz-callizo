1.Cleaning Project and the Datasets

```python
import pandas as pd

# Step 1: Load CSV File
file_path = '/content/drive/MyDrive/Colab Notebooks/Files/SVMtrain.csv'
df = pd.read_csv(file_path)
print("Initial Data:")
print(df.head())

# Step 2: Handle Missing Values
# Separate numeric and non-numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fill missing values for numeric columns with the mean
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing values for categorical columns with a placeholder (e.g., 'Unknown')
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

print("\nMissing values handled:")
print(df.isnull().sum())

# Step 3: Correct Data Types (if needed)
print("\nData types before correction:")
print(df.dtypes)

# Example: Convert categorical column 'Sex' to a category data type
df['Sex'] = df['Sex'].astype('category')

# Example: If needed, convert a column to datetime (not applicable in this dataset)
# df['date_column'] = pd.to_datetime(df['date_column'], errors='coerce')

print("\nData types after correction:")
print(df.dtypes)

# Step 4: Rename Columns for Clarity (optional)
df.rename(columns={
    'PassengerId': 'ID',
    'Pclass': 'Class',
    'SibSp': 'Siblings_Spouses',
    'Parch': 'Parents_Children'
}, inplace=True)

print("\nRenamed Data:")
print(df.head())

# Step 5: Save Cleaned Data (Optional)
output_path = '/content/drive/MyDrive/cleaned_data.csv'
df.to_csv(output_path, index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/drive/MyDrive/Colab Notebooks/Files/cleaned_data.csv'
df = pd.read_csv(file_path)
```

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Histogram for Age
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Age')

# Boxplot for Age
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()

# Histogram for Fare
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Fare'], kde=True, bins=20)
plt.title('Distribution of Fare')

# Boxplot for Fare
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()
```

6. Creation of Heatmaps and of Data and the Numerical Values.

```python
# Bar plot for Survival Rate by Sex
plt.figure(figsize=(12, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.show()

# Bar plot for Survival Rate by Pclass
plt.figure(figsize=(12, 6))
sns.barplot(x='Class', y='Survived', data=df)
plt.title('Survival Rate by Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
print(df.columns)
```

```python
# Calculate the correlation matrix
corr = df[['Age', 'Fare', 'Class', 'Siblings_Spouses',   'Parents_Children']].corr()

# Generate a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

7. Survival Graphics and Rates

```python
# Bar plot for Survival Rate by Gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate statistics by gender
survival_by_gender = df.groupby('Sex')['Survived'].mean()
print("\nSurvival Rate by Gender:")
print(survival_by_gender)
```

```python
# Bar plot for Survival Rate by Class
plt.figure(figsize=(8, 6))
sns.barplot(x='Class', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate statistics by class
survival_by_class = df.groupby('Class')['Survived'].mean()
print("\nSurvival Rate by Passenger Class:")
print(survival_by_class)
```

```python
# Create age bins
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior'])

# Bar plot for Survival Rate by Age Group
plt.figure(figsize=(8, 6))
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title('Survival Rate by Age Group')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate statistics by age group
survival_by_age_group = df.groupby('AgeGroup')['Survived'].mean()
print("\nSurvival Rate by Age Group:")
print(survival_by_age_group)
```
